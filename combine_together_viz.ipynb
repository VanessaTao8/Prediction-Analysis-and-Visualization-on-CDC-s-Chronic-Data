```python
import pandas as pd
```

```python
# load data
combined_data = pd.read_csv('incidence_data.csv')
combined_data = combined_data[combined_data['YearStart'] < 2022]
hospital = pd.read_csv('hospital_number&population_by_states.csv')
```

```python
combined_data.head()
```

```python
hospital.head()
```

```python
hospital = hospital.drop(columns = 'Unnamed: 0')
```

```python
data = combined_data.merge(hospital, left_on = 'LocationAbbr', right_on = 'Location Abbr', how = 'left')
```

```python
data_incidence = pd.read_csv('grey_predict_incidence.csv')
data_incidence = data_incidence[data_incidence['YearStart'] == 2022]
data_incidence = data_incidence[['LocationAbbr', 'Topic', 'DataValueAlt']]
data_incidence.rename(columns = {'DataValueAlt': 'Incidence'}, inplace = True)
```

```python
data_mortality = pd.read_csv('grey_predict_mortality.csv')
data_mortality = data_mortality[data_mortality['YearStart'] == 2022]
data_mortality = data_mortality[['LocationAbbr', 'Topic', 'DataValueAlt']]
data_mortality.rename(columns = {'DataValueAlt': 'Mortality'}, inplace = True)
data_mortality
```

```python
data = data.merge(data_incidence, left_on = ['LocationAbbr','Topic'], right_on = ['LocationAbbr','Topic'], how = 'left')
```

```python
data = data.merge(data_mortality, left_on = ['LocationAbbr','Topic'], right_on = ['LocationAbbr','Topic'], how = 'left')
```

```python
data.to_csv('FinalData.csv')
```
